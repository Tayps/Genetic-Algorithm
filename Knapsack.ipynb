```python
import random as rand
import numpy as np
```

```python
basketSize = 10
solutions = 100
```

```python
# Generate initial Solutions

population = np.random.randint(0,2, (solutions,basketSize))
# population
```

```python
# Create Objects and Solution Containers
limit = {}
value = {}
basket = {}

def createItem(name, lim, val):
    global limit
    global value
    global basket

    limit[name] = lim
    value[name] = val
    basket[name] = int()

createItem('Apple', 50, 5)
createItem('Orange', 60, 4)
createItem('Kiwi', 40, 7)
createItem('Watermelon', 200, 9)
createItem('Grapes', 10, 8)
createItem('Papaya', 150, 1)
createItem('Pear', 55, 10)
createItem('Banana', 70, 11)
createItem('Avocado', 45, 12)
createItem('Dragon Fruit', 100, 4)
```

```python
# Fitness Function
def getFitness():
    global population
    l = list(limit.values()) * population
    v = list(value.values()) * population
    return np.where(np.sum(l, axis = 1) <= 400, 1, 0) * np.sum(v, axis = 1)
    
# getFitness()
```

```python
# Selection Function
def select(pop,fit):
    ix = np.random.choice(solutions, size = solutions, replace=True, p = fit/fit.sum())
    return pop[ix]
```

```python
# Crossover Function
def crossOver(parent, pop):
    if np.random.rand() < 0.8:
        mother = np.random.randint(0,solutions, size =1)
        crossPoint = np.random.randint(0,2,size =basketSize).astype(bool)
        parent[crossPoint] = pop[mother,crossPoint]
    return parent
    
```

```python
# Mutation Function 
def mutate(solution):
    for point in range(basketSize):
        if np.random.rand() < 0.003:
            if solution[point] == 1:
                solution[point] = 0
            else:
                solution[point] = 1
    return solution
```

```python
# Main run
basketSize = 10
solutions = 300

population = np.random.randint(0,2, (solutions,basketSize))

generations = 300
log = {}

for i in range(generations):
    fit = getFitness()
    population = select(population, fit)
    log[i] = np.average(fit)
    
    for parent in population:
        popCopy = population.copy()
        child = crossOver(parent, popCopy)
        child = mutate(child)
        parent[:] = child
    
    # print(i)
    # print(population, np.average(fit))
```
